# **Case Study – Shopping Cart System**
**Instructions:**
1. Create a list named cart with initial items: Milk, Bread, Eggs, Apple
2. Perform list operations: • Add multiple items using extend() • Insert item at index 2 using insert() • Remove item Apple using remove() • Remove last item using pop()
3. Use list methods: • index() to find position of Eggs • count() to count occurrences of Milk
4. Apply indexing: • Display first item • Display last item
5. Apply slicing: • Display first 3 items from the list
6. Find total number of items using len()
7. Display all outputs using print()

```python
# Shopping Cart System

#  Create cart list
cart = ["apple", "bread", "eggs", "milk"]

# List operations
# add multiple item
cart.extend(["butter", "cheese"])
cart.insert(2, "juice")
cart.remove("apple")
cart.pop()

# List methods
egg_position = cart.index("eggs")
milk_count = cart.count("milk")

# Indexing
first_item = cart[0]
last_item = cart[-1]

# Slicing
first_three_items = cart[:3]

# Total items
total_items = len(cart)

# Display outputs
print("Final Cart:", cart)
print("Position of Eggs:", egg_position)
print("Count of Milk:", milk_count)
print("First Item:", first_item)
print("Last Item:", last_item)
print("First 3 Items:", first_three_items)
print("Total Items:", total_items)
```
