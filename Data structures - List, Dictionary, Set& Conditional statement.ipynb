Data structures - List, Dictionary, Set& Conditional statement

```python
#list

#list creation

age_list = [24, 25, 27, 28, 29]

name_list = ["Akash", "Arun", "Kavi", "Nila", "Ram"]

print("Original age_list:", age_list)
print("Original name_list:", name_list)

#list operation/ modification

# Append
name_list.append("Yazhini")
print("\nAfter Append:", name_list)

# Insert
age_list.insert(2, 26)
print("After Insert:", age_list)

# Remove
name_list.remove("Yazhini")
print("After Remove:", name_list)

# Pop last element
age_list.pop()
print("After Pop:", age_list)

# Extend
age_list.extend([31, 30, 32])
print("After Extend:", age_list)

# Sort descending
age_list.sort(reverse=True)
print("Descending Order:", age_list)

# Max, Min, Sum
print("Maximum Age:", max(age_list))
print("Minimum Age:", min(age_list))
print("Sum of Ages:", sum(age_list))


#Accessing List Elements

# First element
print("\nFirst Element:", name_list[0])

# Last element
print("Last Element:", name_list[-1])

# Elements index 2 to 4
print("Index 2 to 4:", name_list[2:5])

# Reverse order
print("Reverse Order:", name_list[::-1])
```

```python
 # DICTIONARY

 # Create dictionary
student_marks = {
    "Akash": 78,
    "Arun": 85,
    "Kavi": 90,
    "Nila": 88,
    "Ram": 75
}

print("\nStudent Marks:", student_marks)

# Access specific student mark
print("Akash Mark:", student_marks["Akash"])

# Add new student
student_marks["Janani"] = 80
print("After Adding Janani:", student_marks)

# Update mark
student_marks["Ram"] = 82
print("After Updating Ram:", student_marks)

#  keys,values,items
print("Keys:", student_marks.keys())
print("Values:", student_marks.values())
print("Items:", student_marks.items())
```

```python
#set

# Create set
my_set = {'a', 'e', 'i', 'o', 'u', 'a', 'a', 'i'}

print("\nMy Set:", my_set)

print("Explanation:")
print("Duplicate values are automatically removed in sets.")

# Access my_set[4]
print("\nSets do not support indexing.")
print("Trying my_set[4] gives TypeError.")

# Create two sets
set1 = {1, 3, 5, 7, 9}
set2 = {2, 3, 5, 8, 10}

# Union and Intersection
print("Union:", set1.union(set2))
print("Intersection:", set1.intersection(set2))
```

```python
#conditional statements
score = int(input("Enter your score (0 to 10): "))

if score < 0 or score > 10:
    print("Invalid Score! Please enter between 0 and 10.")

elif score > 7:
    print(" Excellent performance! Keep it up.")

elif score >= 4 and score <= 7:
    print(" Good effort! Keep practicing, there's room for improvement.")

elif score >= 0 and score <= 4:
    print(" Need to improve your performance; consistent practice will lead to better results.")

else:
    print("Invalid Score! Please enter between 0 and 10.")
```
